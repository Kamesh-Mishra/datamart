```python
import requests
```

```python
headers = {'Accept': 'application/json'}
```

```python
r = requests.get('http://catalog.data.gov/api/3/action/current_package_list_with_resources',
                 headers=headers, params=dict(rows=1000, start=302000))
r.raise_for_status()
obj = r.json()
```
